```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
sns.set_theme(style="white",context="talk")

# Set up the matplotlib figure
f, (ax1) = plt.subplots(1, 1, figsize=(12, 8) , dpi=144)

# Generate some sequential data
y1 = np.array([55, 60, 65, 69, 73, 76, 79, 81, 84])
bar1 = np.array([56, 60, 66, 71, 75, 78, 81, 83, 86])
pie1 = np.array([64, 66, 68, 71, 73, 75, 77, 79, 80])
line1 = np.array([41, 45, 50, 54, 59, 62, 66, 69, 72])
scatter1 = np.array([46, 51, 57, 62, 67, 72, 76, 79, 82])
x1 = np.array([0, 1.25, 2.5, 3.75, 5, 6.25, 7.5, 8.75, 10])

data=pd.DataFrame(index=x1, data={'average':y1, 'bar': bar1, 'pie': pie1, 'line': line1, 'scatter': scatter1})


sns.lineplot(data=data, ax=ax1, markers=True, dashes=False)
#ax1.axhline(0, color="k", clip_on=False)
ax1.set(ylim=(20, 100))
#ax1.set(xlim=(0, 10))
ax1.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax1.set_xlabel("Fovea region")
ax1.set_ylabel("Fixations on AOI (%)")

# Finalize the plot
#sns.despine(bottom=True)
plt.setp(f.axes, yticks=[40, 50, 60, 70, 80])
#plt.tight_layout(h_pad=2)


f.savefig('FixationCoverage.png')
```
