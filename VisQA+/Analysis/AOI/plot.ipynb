```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
sns.set_theme(style="white",context="talk")

# Set up the matplotlib figure
f, (ax1) = plt.subplots(1, 1, figsize=(12, 8) , dpi=144)

# Generate some sequential data
y1 = np.array([55, 61, 66, 70, 74, 77, 81, 84.7, 88])
bar1 = np.array([56, 62, 68, 73, 77, 80, 85, 87, 91])
pie1 = np.array([64, 67, 71, 74, 77, 79, 83, 85, 88])
scatter1 = np.array([46, 52, 58, 63, 68, 72, 77, 82, 86])
line1 = np.array([41, 46, 51, 55, 59, 63, 67, 71, 76])
x1 = np.array([0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.4, 0.5])

data=pd.DataFrame(index=x1, data={'average':y1, 'bar': bar1, 'pie': pie1, 'scatter': scatter1, 'line': line1})


sns.lineplot(data=data, ax=ax1, markers=True, dashes=False)
ax1.set(ylim=(20, 100))
#ax1.set(xlim=(0, 10))
#ax1.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax1.set_xlabel("Fovea region")
ax1.set_ylabel("Fixations on AOI (%)")

# Finalize the plot
sns.despine()
plt.setp(f.axes, yticks=[40, 50, 60, 70, 80])
#plt.tight_layout(h_pad=2)


f.savefig('FixationCoverage.png')
```
