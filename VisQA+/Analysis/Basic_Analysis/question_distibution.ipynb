## üìù Questions distribution analysis ##


```python
import json
import pandas as pd
import os
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
directory = r'C:\Users\karth\Desktop\Karthik\HCI_Lab\merged\qa'
df = pd.DataFrame()
list_of_df = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    
    # checking if it is a file
    if os.path.isfile(f):
        json_file = open(f)
        qa_file = json.load(json_file)
        visualization_type = qa_file['vistype']
        for i in range(1,6):
            type_of_question = qa_file['Q'+str(i)]['type']
            answer = qa_file['Q'+str(i)]['answer']
            update_entry = {'vistype':visualization_type,'type':type_of_question,'answer':answer}
            list_of_df.append(update_entry)
df = df.append(pd.DataFrame(list_of_df))
```

## üìà Plotting Visualization type distribution ##

```python
distibution_of_vistype = df['vistype'].value_counts().to_frame().reset_index()
distibution_of_vistype.columns = ['visualization_type','Number of visualizations']

vistype_distirbution =sns.barplot(data=distibution_of_vistype, x='visualization_type', y='Number of visualizations',)
vistype_distirbution.set(xlabel='visualization_type', ylabel='Number of visualizations in all images',
       title='visualization_type distribution in all the images')

# Show the plot
plt.show()

vistype_distirbution.figure.savefig('vistype.png', dpi=600)
```

## ‚ùì Plotting the question type ##

```python
distibution_of_question_type = df['type'].value_counts().to_frame().reset_index()
distibution_of_question_type.columns = ['question_type','Number of questions']

question_type_distirbution =sns.barplot(data=distibution_of_question_type, x='question_type', y='Number of questions',)
question_type_distirbution.set(xlabel='question_type', ylabel='Number of questions in all images',
       title='question_type distribution in all the images')

# Show the plot
plt.show()

question_type_distirbution.figure.savefig('vistype.png', dpi=600)
```
