```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# assign directory
directory = r'C:\Users\karth\Desktop\Karthik\HCI_Lab\VisQA\element_labels'
count = pd.Series()
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)

    # checking if it is a file
    if os.path.isfile(f):
        df=pd.read_csv(f, header=None)
        df = df.drop_duplicates(subset=[0]).reset_index()
        count = count.append(df[1].value_counts())
  
    
```

```python
df = count.to_frame()
df = df.reset_index()
df = df.rename(columns = {'index':'label', 0:'count'})
# count.groupby('Column1')['Column2'].sum()
element_counts_sum = df.groupby('label')['count'].sum()
element_counts_sum = element_counts_sum.to_frame().reset_index()
element_counts_sum= element_counts_sum.sort_values(by=['count'])
```

 # üìà Plot the Distribution

```python
Elemenet_label_distirbution =sns.barplot(data=element_counts_sum, x='label', y='count',)
Elemenet_label_distirbution.set_xticklabels(labels = element_counts_sum.label, rotation=90)
Elemenet_label_distirbution.set(xlabel='Element_labels', ylabel='Number of Labels in all images',
       title='Element label distribution in all the images')

# Show the plot
plt.show()

Elemenet_label_distirbution.figure.savefig('labels.png', dpi=600)
```
